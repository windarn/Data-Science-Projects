### Generación de numeros aleatorios

```python
import numpy as np
```

```python
## Generar un numero aleatorio entero entre (1,100)
np.random.randint(1,100)
```

```python
##La forma más clásica de generar un npumero aleatorio es entre 0 y 1 (con decimales)
np.random.random()
```

```python
##Función que genera una lista de n números aleatorios enteros dentro del intervalo [a,b]
def randint_list(n, a, b):
    x = []
    for i in range(n):
        x.append(np.random.randint(a,b))
    return x
```

```python
randint_list(25, 1, 50)
```
